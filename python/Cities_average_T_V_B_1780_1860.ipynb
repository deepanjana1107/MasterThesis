```python
import numpy as np 
import pandas as pd
```

```python
vienna=pd.read_csv('/Users/DMac/OneDrive/MasterThesis/data/HISTALP_AT_WIE_T01_1780_1860_Clean.csv', sep=",")
vienna = vienna.set_index('year')
bud=pd.read_csv('/Users/DMac/OneDrive/MasterThesis/data/HISTALP_HU_BUD_T01_1780_1860_Clean.csv', sep=",")
bud=bud.set_index('year')
```

```python
cities_concat = pd.concat((vienna, bud))
by_row_index = cities_concat.groupby(cities_concat.index)
cities_means = by_row_index.mean()
cities_means_adjusted = cities_means/10.0
print cities_means_adjusted.head()
```
